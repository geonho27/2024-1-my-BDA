#mpg 통합 연비 변수 만들기

```python
mpg = pd.read_csv('data/mpg.csv')
mpg.sample(8)
```

```python
# 통합 연비 변수(total) 만들기, total <= cty + hmy의 평균: 시험 예상
mpg['total'] = (mpg['cty'] + mpg['hwy']) / 2
mpg.head()
```

```python
#통합 연비 변수(mpg_mean) 만들기, total <= cty + hwy으; 평균: 시험 예상
mpg['mpg_mean'] = mpg[['cty', 'hwy']].mean(axis='columns')
mpg.head()
```

```python
sum(mpg['total']) / len(mpg) #total 합계를 행 수로 나누기
```

```python
mpg['total'].mean() # 통합 연비 변수 평균
```

# 중첩 조건문 활용하기

```python
mpg['grade'] = np.where(mpg['total']>= 30, 'A',
               np.where(mpg['total']>= 20, 'B', 'C'))

mpg.sample(5)
```

```python
count_grade = mpg['grade'].value_counts()
count_grade
```

```python
count_grade.plot.bar(rot = 0)
```

```python
sns.countplot(mpg, x="grade", hue="grade", stat='percent')
```

```python
mpg['grade'].value_counts().sort_index().index
```

# 정리하기

```python
#1. 패키지 로드
import pandas as pd
import numpy as np

# 2. 데이터 불러오기
mpg = pd.read_csv('mpg.csv')

#3. 데이터 파악하기
mpg.shape
mpg.head()
mpg.tail()
mpg.info()
mpg.describe()

# 4. 변수명 바꾸기
mpg = mpg.rename(columns = {'manufactuer' : 'company'})

# 5. 파생변수 만들기
mpg['total'] = (mpg['cty'] + mpg['hwy']) / 2
mpg['test'] = np.where(mpg['total']>=20, 'pass', 'fail')

# 6. 빈도 확인하기
count_test = mpg['test'].value_counts()
count_test.plot.bar(rot = 0)
```
